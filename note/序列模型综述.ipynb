## 序列数据
1、电影评分。 电影的评价随时间变化而变化  
2、音乐、语言、文本、视频都是连续的  
3、大地震发生后，很可能会有几次较小的余震  
4、人的互动是连续的  
5、预测明天的股份要比填补昨天遗失的股份更难  

## 序列数据建模
在时间 t 观察到 x_t，可得到 T 个不独立的随机变量  
`(x_1,...x_T)~p(x)`  

联合概率使用条件概率展开  
`p(a,b)=p(a)p(b|a)=p(b)p(a|b)`  

序列模型中，当前数据跟之前观察到的数据相关  


#### 对条件概率建模
对见过的数据建模，也称 `自回归模型`，自回归模型使用自身过去数据来预测未来  


## 方案1：马尔科夫假设
假设当前数据只跟最近 `τ` 个过去数据点相关，从而简化模型  


## 方案2：潜变量模型
引入潜变量 `h_t` 来概括历史信息  
RNN 属于潜变量模型  
`h_t = f(x1,...,x_t-1)`，`x_t = p(x_t|h_t)`
