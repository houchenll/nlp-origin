## 初识 numpy
numpy 其实就是一个多维的数组对象

```python
import numpy as np

data = [1,2,3,4,5]
n = np.array(data * 10)
print(n)
```

每一个numpy的数组都有一个shape和一个叫dtype的属性

```python
n.shape  # shape表示获取获取到np数组的维度
```

```python
n.dtype    # dtype表示获取到数组的类型
```

## 嵌套序列
定义：由一组等长列表组成的列表

```python
arr = [[1,2,3,4], [1,2,3,4]]
print(arr)
arr2 = np.array(arr)
print(arr2)
print(arr2.ndim)    # ndim 表示外层维度
print(arr2.shape)    # shape 表示完整维度：2行4列
```

## np对数据类型的一个判断
数组中同时有字符串和数字时，np自动把数字转成字符串  
当成员中有1个是float时，np自动把所有数字转成float  
int64 -> float64 -> unicode  

### numpy中数据类型
int类型：int8, int16, int32, int64  
float类型：float16, float32, float64, float128

```python
arr = [['1','2',3,4],[5,6,7,8]]
arr2 = np.array(arr)
print(arr2)
print(arr2.dtype)   # U表示unicode
```

```python
arr = [[1,2,3,4],[5,6,7,8]]
arr2 = np.array(arr)
print(arr2)
print(arr2.dtype)
```

```python
arr = [[1.1,2,3,4],[5,6,7,8]]
arr2 = np.array(arr)
print(arr2)
print(arr2.dtype)
```
