# 矢量化
矢量化：数组不用在编写循环的情况下就可以进行批量运算

## 对应位置相加/减/乘/除

```python
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
arr1 + arr2
```

```python
arr1 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
arr2 = np.array([[5, 6, 7, 8], [5, 6, 7, 8]])
arr1 + arr2
```

```python
arr1 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
arr2 = np.array([[5, 6, 7, 8], [5, 6, 7, 8]])
arr2 - arr1
```

```python
arr1 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
arr2 = np.array([[5, 6, 7, 8], [5, 6, 7, 8]])
arr1 * arr2
```

```python
arr1 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
arr2 = np.array([[5, 6, 7, 8], [5, 6, 7, 8]])
arr1 / arr2
arr2 / arr1
```

```python
arr1 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
5 * arr1
```
