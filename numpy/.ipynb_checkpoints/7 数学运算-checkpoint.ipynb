常用函数：sum, mean, std, argmin, argmax, cumsum

## 对所有元素的运算

```python
import numpy as np

arr = np.random.randn(4,4)
print(arr)
```

```python
print(arr.mean())  # 所有值的平均值
print(np.mean(arr))  # 同上

print(arr.sum())  # 所有元素求和
print(arr.std())  # 所有元素求标准差
```

## 在一个轴上运算

```python
print(arr.mean(axis=1))  # 计算轴1上的平均值
print(arr.sum(0))  # 计算轴0上的和
```

## 排序方法

```python
arr = np.random.randn(4)
print(arr)
```

```python
arr.sort()  # 从小到大
print(arr)
```

多维数组排序

```python
arr = np.random.randn(4,4)
print(arr)
#arr.sort(1)  # 传递轴号，按轴排序
#print(arr)
arr.sort()  # 按行排序
print(arr)
```
