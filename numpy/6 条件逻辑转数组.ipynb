np.where，等于同python中 x if condition else y
np.where 一般用于根据一个数组生成一个新的数组，跟Python中map/reduce等函数类似

```python
import numpy as np

x_arr = np.array([1.1,1.2,1.3])
y_arr = np.array([2.1,2.2,2.3])
condition = np.array([True, False, True])
result = [(x if c else y) for x,y,c in zip(x_arr, y_arr, condition)]   # zip接收的参数是可迭代的对象，用于将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的列表
print(result)
r = np.where(condition, x_arr, y_arr)
print(r)
```

## 值替换

```python
arr = np.random.randn(4,4)
print(arr)
arr_1 = np.where(arr>0, 2, -2)
print(arr_1)
arr_2 = np.where(arr>0, 2, arr)
print(arr_2)
```
