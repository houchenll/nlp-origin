### 去空格及特殊字符

```python
s = ' \n\thello, world!'
print(s.strip())  # 去除字符串开头和结尾的空格、\t、\n
print(s.lstrip(' hello, '))  # 去除字符串开头指定的字符串
print(s.rstrip('!'))
```

### 连接字符串

```python
s1 = 'strcat'
s2 = 'append'
s1 += s2
print(s1)
```

### 查找字符

```python
flag = 'r'
pos = s1.index(flag)  # index表示字符第一次出现的位置，< 0 为未找到
print(pos)
```

### 查找字符串

```python
s1 = 'abcdefg'
s2 = 'cde'
pos1 = s1.find(s2)  # s2在s1中第一次出现时的起始位置
print(pos1)
pos2 = s1.index(s2)  # 同find
print(pos2)
```

### 计算字符串中出现频次最多的字母

```python
# version 1
import re
from collections import Counter

def get_max_value_v1(text):
    text = text.lower()
    result = re.findall('[a-z]', text)  # 去掉列表中的非字母
    count = Counter(result)  # 统计各个字母出现的次数 Counter({'l': 3, 'o': 2, 'd': 1})
    count_list = list(count.values())
    max_value = max(count_list)
    max_list = []
    for k, v in count.items():
        if v == max_value:
            max_list.append(k)
    max_list = sorted(max_list)
    return max_list[0]
```

```python
get_max_value_v1('hello, world!')
```

```python
# version 2

def get_max_value_v2(text):
    count = Counter([x for x in text.lower() if x.isalpha()])  # 遍历text，得到各个字符x，如果符合if条件，就返回x
    m = max(count.values())
    return sorted([x for (x, y) in count.items() if y == m])[0]  # 同上
```

```python
get_max_value_v2('hello, world!')
```

```python
# version 3
import string

def get_max_value_v3(text):
    text = text.lower()
    # count("x")函数会返回每个字母出现几次
    # 把每个字母都应用count()，并返回最大那个字母
    return max(string.ascii_lowercase, key=text.count)
```

```python
get_max_value_v3('hello, world!')
```
